# Imports

```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import QTable, vstack, join
```

## Shared IP Distribution

```python
nasa_sum = QTable.read('data/nasa_ipcum.ecsv', format='ascii.ecsv')
nasa_pal = nasa_sum['ps_cum'].astype(float)
```

```python
esa_sum = QTable.read('data/esa_ipcum.ecsv', format='ascii.ecsv')
esa_pal = esa_sum['ps_cum'].astype(float)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

bins = np.arange(-14, 0, 0.25)

ax.hist(nasa_pal, bins=bins, alpha=0.5, label='NASA', color='orange')
ax.hist(esa_pal, bins=bins, alpha=0.5, label='ESA', color='blue')

ax.set_title("Palermo Scale Distribution")
ax.set_xlabel("Palermo Scale")
ax.set_ylabel("Asteroid Counts")

ax.legend()

plt.show();
```
