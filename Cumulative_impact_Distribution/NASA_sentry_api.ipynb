# Importing NASA NEO dataset by reading json into csv

```python
import sys
print(sys.executable)
import requests
import datetime
import json

# NASA's api (summary)
response = requests.get("https://ssd-api.jpl.nasa.gov/sentry.api")

# Creates a json file out of the data pulled with NASA's api
if response.status_code == 200:
    data = response.json()

# Saves the json file to this directory
    with open('NASA_download.json', 'w') as f:
        json.dump(data, f, indent=4, sort_keys=True)
    print("Data saved!")
else:
    print("Failed. {response.status_code}")

# Gives the date of the download so that you can tell if the file is up to date with the current database
timestamp = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
filename = f"NASA_download_{timestamp}.json"
```

```python
import json
import csv

# Loads the json file saved above
with open('NASA_download.json') as jf:
    loaded_json = json.load(jf)
    
json_data = loaded_json['data']

# Creates a csv called 'NASA_data.csv'
new_csv = open('NASA_data.csv', 'w')

# Sets up to begin writing in the new csv
writing = csv.writer(new_csv)

# Setting a counter to count up for every written index in the csv
counter = 0

# Writing the csv iteratively
for key in json_data:
    if counter == 0:

        # Writing headers of the csv according to the keys in the json
        column = key.keys()
        writing.writerow(column)
        counter += 1

    # Writing data of csv file according to the values in the json
    writing.writerow(key.values())

new_csv.close()
```
